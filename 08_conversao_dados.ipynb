# Tipo de Dados e Conversão
Tipo Inteiro (int)

Ponto Flutuante ou Decimal (float)

String (str)

String (Object)

Tipo de dado é a forma como o computador entende e armazena a informação. Cada tipo de dado define que operações podem ser feitas com aquela informação e quanto espaço ela ocupa na memória.



- Tipos de dados dizem ao computador o que cada valor representa.

- Servem para realizar operações corretas e otimizar o uso da memória.

- Converter tipos é necessário quando queremos usar os dados de forma diferente do tipo original, evitando erros e garantindo precisão nos cálculos.


```python
import pandas as pd
```

```python
# Ler um arquivo CSV é similar ao arquivo xlsx, a diferença é , delimiter=',' que indica que o arquivo é separado por vírgula.
df = pd.read_csv(r"C:\Users\Jailson\Downloads\Eu. Python do Básico ao Avançado\list\analytics-with-python\arquivo csv\Clientes.csv", delimiter=',') 
df
```

# Conhecendo os tipos de Dados

Tipo de dados de colunas especificas

```python
#O que ele faz:

    # df é um DataFrame do pandas (uma estrutura de dados usada em Python para trabalhar com tabelas, tipo Excel).

    # df['street'], df['cell_phone'] e df['id'] selecionam colunas específicas do DataFrame.

    # .dtype mostra o tipo de dado da coluna selecionada.

    # print() exibe o tipo no console.

print(df['street'].dtype)
print(df['cell_phone'].dtype)
print(df['id'].dtype)
```

Tipo de Dados de todas as Colunas 

```python
for colunas in df.columns:
    print(f"Coluna: {colunas}. Tipo de Dado: {df[colunas].dtype}")
```

#### Conversão de Dados

```python
# Converter tudo para string
df['cell_phone'] = df['cell_phone'].astype(str)

# Remover tudo que não for número
df['cell_phone'] = df['cell_phone'].str.replace(r'\D', '', regex=True)

# Converter para inteiro (permitindo nulos)
df['cell_phone'] = df['cell_phone'].replace('', pd.NA).astype('Int64')

```
