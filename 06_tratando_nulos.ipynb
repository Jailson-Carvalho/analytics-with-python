A forma de substituição depende do tipo de dado e do objetivo: excluir, preencher com outros valores

#### Substituir valores NaN usando o fillna. Usei de exemplo a coluna email, contudo o mesmo método se aplica a todas as colunas. Veja a explicação abaixo.

- df['email']

    Seleciona a coluna chamada email do DataFrame df.

- .fillna('Não Cadastrado')

    O método fillna() serve para substituir valores nulos (NaN).

    Nesse caso, todo valor faltante (nulo) na coluna email será trocado por 'Não Cadastrado'.

- df['email'] = ...

    O resultado da substituição é atribuído de volta à mesma coluna email.

    Isso significa que o DataFrame df agora terá a coluna email sem valores nulos.

```python
df['email'] =           df['email'].fillna('Não Cadastrado')
df['cell_phone'] =      df['cell_phone'].fillna('Sem Registro')
df['street'] =          df['street'].fillna('Sem Cadastro')
df['number'] =          df['number'].fillna('Sem Número')
df['additionals'] =     df['additionals'].fillna('Sem Informações')
```

```python
df.head(10)

# Veja que no resultado abaixo os valores NaN foram devidamente substituidos.
```

```python
# substituir valores dentro da tabela
# replace faz essa substituição

df['state'] = df['state'].replace('Bahia', 'BA')
# variável contendo a coluna que quero: df['state']
# recebe o Dataframe com a coluna: df['state']
# recebe o comando que faz a alteração: .replace
# recebe como está: ('Bahia'
# e como vai ficar: , 'BA')
df
```

```python
# mesmo resultado usando variáveis
original = 'São Paulo'
subistituto = 'SP'

df['state'] = df['state'].replace(original, subistituto)

df.head(5)
```

```python
# modificar varios de uma só vez usando o mapeamento
# usaremos um dict
mapeamento = {

    'Acre': 'AC',
    'Rio Grande do Sul': 'RS',
    'Mato Grosso': 'MG'

}

df['state'] = df['state'].replace(mapeamento)

df['state'] =           df['state'].fillna('Não Cadastrado')


df
```
